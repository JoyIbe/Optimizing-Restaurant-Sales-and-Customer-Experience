# Key questions to address

### Top 5 restaurants by Revenue

```python
#Top 5 restaurants by Revenue

df.groupby('Name')['Revenue'].sum().sort_values(ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Visualizing the top 5 restuarant

restuarant=df.groupby('Name')['Revenue'].sum().sort_values(ascending=False).head(5)

plot = sns.barplot(data=restuarant, palette=sns.color_palette('tab10'))
for bars in plot.containers:
    plot.bar_label(bars, [f'{x:.2f}' for x in bars.datavalues])
plot.set_xlabel('Restuarants')
plot.set_ylabel('Total Revenue')   
plt.title('Top 5 restaurants by Revenue')
plt.xticks(rotation=45, ha='right')
plt.show

# Outcome: 
print("""
Insight:The highest-earning restaurant is Restaurant 4324, with a revenue of about $1,531,867.55.
The other four restaurants follow closely behind, with revenues ranging from $1,509,391.07 to $1,518,965.19.""")
```

### 5 most rated restaurants

```python
# Grouping restuarants by the highest ratings on average

df.groupby('Name')['Rating'].mean().sort_values(ascending=False).head(5)
```

```python
# Visualizing 

rated_restuarant = df.groupby('Name')['Rating'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(20, 7)) # manipulating its size

plt.bar(rated_restuarant.index, rated_restuarant.values)
plt.xlabel('Name')
plt.ylabel('Average Rating')
plt.title('5 most rated restuarants by Average Rating', fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.show()

# Outcome: 
print('Insight:Restaurants with the most highest ratings on average is Restaurant 7500, Restuarant 7744,7680, 7682 with all having a ratings of 5.0')
```

### Carry out EDA to find the relation between revenue and features of the dataset.

```python
# Calculating the correlation between rating and revenue generated

correlation_matrix = df[['Revenue', 'Rating']].corr()
correlation_matrix 
```

```python
# Visualizing the correlation between rating and revenue generated

correlation_matrix = df[['Revenue', 'Rating']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1.5)
plt.title("Correlation Between Revenue and Ratings")
plt.show()

print("""
Insight: There is no relationship between a restaurantâ€™s revenue and its rating.
This indicates that high ratings do not necessarily lead to high revenue.
""")
```

```python
# Calculating locations that generates the most revenue

df.groupby('Location')['Revenue'].sum().sort_values(ascending=False)
```

```python
# Visualizing locations that generates the most revenue

location=df.groupby('Location')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))

plot = sns.barplot(data=location, palette=sns.color_palette('tab10'))
for bars in plot.containers:
    plot.bar_label(bars, [f'{x:.2f}' for x in bars.datavalues])
plot.set_xlabel('Location')
plot.set_ylabel('Total Revenue')   
plt.title('Top Location that generates the Highest Revenue')
plt.xticks(rotation=45, ha='right')
plt.show

# Outcome: 
print("""
Insight: Downtown is the highest generating location and shows a significant gap compared to Suburban and Rural regions. 
Leveraging on this location will boost sales and revenue growth and also improves brand visibility. 
""")
```

```python
# Calculating Cuisine that generates the most revenue

df.groupby('Cuisine')['Revenue'].sum().sort_values(ascending=False)
```

```python
# Visualizing Cuisine that generates the most revenue

cuisine = df.groupby('Cuisine')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))

plot = sns.barplot(data=cuisine, palette=sns.color_palette('tab10'))
for bars in plot.containers:
    plot.bar_label(bars, [f'{x:.2f}' for x in bars.datavalues])
plot.set_xlabel('Cuisine')
plot.set_ylabel('Total Revenue')   
plt.title('Top Cuisine that generates the Highest Revenue')
plt.xticks(rotation=45, ha='right')
plt.show

# Outcome: 
print('Insight: Japanese cuisine leads with a total revenue, french, italian follows closely. This shows highly profitable cuisines choices.')
```
